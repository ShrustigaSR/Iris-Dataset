# Iris

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris=pd.read_csv("iris.csv")
```

```python
iris.head()
```

```python
#Bar Plot
sns.barplot(x='Species',y='SepalLengthCm',data=iris)
```

```python
#Box Plot
plt.style.use('ggplot')
plt.subplot()
sns.boxplot(x='Species',y='SepalWidthCm',data=iris)
```

```python
#Joint Plot
sns.jointplot(x='PetalLengthCm',y='PetalWidthCm',data=iris[0:20])
```

```python
#Scatter Plot

sns.scatterplot(x="SepalLengthCm",y="SepalWidthCm",data=iris)
```

```python
#Pair Plot
sns.pairplot(iris[0:10])
```

```python
#KDE Plot - Kernal Distribution Estimation
sns.kdeplot(iris['PetalWidthCm'], shade=True)
```

# Tips

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips=pd.read_csv('tips.csv')
tips.head()
```

```python
#Bar Plot
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
#Box Plot
plt.style.use('ggplot')
plt.subplot()
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
#Joint Plot
sns.jointplot(x='total_bill',y='tip',data=tips[0:20])
```

```python
#Scatter Plot

sns.scatterplot(x="total_bill",y="tip",data=tips)
```

```python
#Pair Plot
sns.pairplot(tips[0:10])
```

```python
#KDE Plot
sns.kdeplot(tips['total_bill'], shade=True)
```
